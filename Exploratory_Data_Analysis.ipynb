### Population Density
We believed population density was an important factor to consider in analyzing the spread of COVID-19 because transmission is person to person through close (less than 6 feet) of contact. After pulling the population density in people per square mile from the US Census, we found the average population density to be 198 people per square mile and the median to be 103 people/sqmi. Note that this figure is different than the overall population density in the United States (94 people per square mile) since we are averaging state values and not the total population and land area. With that being noted, the least populous state in our data set was Alaska (1 person per square mile) and the most populous was New Jersey (1210 people per square mile).

### Median Household Income
We believed household income was an important factor to consider in analyzing the spread of COVID-19 because behaviors, from occupation to dietary habits, may be different among people of different income brackets. After pulling the median income data from the Federal Reserve, we found the average state's median income \$60,181 and the median to be \$58,848. Note that this figure is different than the overall median income in the United States (\$63,688/year) since we are averaging state values and not the income and population. With that being noted, the least wealthy state in our data set was Mississippi (\$43,567/ year) and the most populous was Maryland (\$81,868/year).

### Median Age
We believed age was an important factor to consider in analyzing the spread of COVID-19 because preliminary studies have shown that older people are generally more susceptible and more likely to die from the disease. After pulling the median age data from the US Census, we found the average state's median age is 38.6 years and the median to be 38.55 Note that this figure is different than the overall median age in the United States (38.2 years) since we are averaging state values and not the income and population. With that being noted, the state with the oldest average in our data set was Maine at 44.9 years and the youngest average was Utah at 31. 

### Physicians Per 1000 People
We believed physicians per 1000 people was an important factor to consider in analyzing the spread of COVID-19 because we had a suspicion that healthcare workers would reduce the spread of disease and the likelihood that infected patients would die from it. After pulling active physicians from the Kaiser Family Foundation and dividing it by each state's population, we found the average number of physicians per 1000 people to be 6.91 per 1000 and the median to be 2.88 physicians per 1000. The state with the least physicians is Ohio with 0.17 per 1000, and the state with the highest is Rhode Island with 48.30 per 1000. 

### Cases of COVID-19 Per 1000 People
We believed this figure was important because it was a leading indicator for the number of individuals hospitalized or dead. Also, this is the most basic status/ form of COVID-19 infection, thus we believed this would be critical in our data analysis. New York is the hardest hit state with 9.65 cases per 1000 people while Minnesota is the least impacted with 0.29 per 1000.

### Percentage of Population in Urban Areas 
The COVID-19 virus spreads through droplets that are generated when someone who is infected coughs or sneezes. Urban areas have a higher concentration of people that other collections of the population, so the virus spreads faster in these places. Knowing the percentage of the population in an Urban area is paramount is understanding how susceptible an area is to spread of the virus. The average of all  the states percentage of the population who lives in urban area is 75%. The state with the least urban population is Maine at 38.70%. The state with the higest urban population is New York at 95%. 

### Average Annual Temperature 
While the weather is not directly responsible for infecting others, cold temperature allows the virus to be spread more easily. Therefore knowing average tempearture could be indicative of predicting which geographies COVID-19 would spread faster in. The state with the coldest temperature is Alaska (-3 degrees Celsius) and the state with the hottest temperature is Florida (21.5 degrees Celsius) 

### Percent Smokers by State
Smokers are likely to be more vulnerable to COVID-19 as the act of smoking means that fingers are in contact with lips which increases the possibility of transmission of virus from hand to mouth. Smokers may also already have lung disease or reduced lung capacity which would greatly increase risk of serious illness. Therefore States with more smokers would likely see more deaths of COVID-19. The state with the highest percentage of smokers is Wyoming at 26% and the state with the lowest percentage of smokers is Alabama at 8.9%


```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
```

```
import io 
df = pd.read_excel(io.BytesIO(uploaded['covid19_Project_states_data.xlsx']))
```

```
#separate states into two groups
first_half = df.head(25)
second_half = df.tail(n=25)

# Vizualize distribution of Physicians Per 1000 People by State
first_half.plot(kind='bar',x='State',y='Physicians Per 1000 People (calculation)', color = 'red', figsize=(30,10))
second_half.plot(kind='bar',x='State',y='Physicians Per 1000 People (calculation)', color = 'blue', figsize=(30,10))

# Vizualize distribution of Positive Cases by State
first_half.plot(kind='scatter',x='State',y='Number of Positive Cases (as of 4/13/2020) Kaggle', color = 'red', figsize=(30,10))
second_half.plot(kind='scatter',x='State',y='Number of Positive Cases (as of 4/13/2020) Kaggle', color = 'blue', figsize=(30,10))

#Vizualize distribution of Median Household Income per Year ($) by State
first_half.plot(kind='barh',x='State',y='Median Household Income $/year (US Census Bureau)', color = 'red', figsize=(30,10))
second_half.plot(kind='barh',x='State',y='Median Household Income $/year (US Census Bureau)', color = 'blue', figsize=(30,10))
```
